#Generation of Confidence Factor Matrix

```python
def getscore(minvalue, maxvalue, value):
  average = (minvalue+maxvalue)/2
  if value >= minvalue and value <= average:
    return (value-minvalue)/(average-minvalue)
  if value >= average and value <= maxvalue:
    return (maxvalue-value)/(maxvalue-average)
  else:
    return 0

confidence_factor_array = []
for i in range(0,len(test)):
  arr=[]
  for j in range(0,4):
    if (j==0):
      fuzzy_rank_setosa = getscore(min_sepal_length_setosa, max_sepal_length_setosa, test.loc[i][j])
      fuzzy_rank_versicolor = getscore(min_sepal_length_versicolor, max_sepal_length_versicolor, test.loc[i][j])
      fuzzy_rank_virginica = getscore(min_sepal_length_virginica, max_sepal_length_virginica, test.loc[i][j])
    if (j==1):
      fuzzy_rank_setosa = getscore(min_sepal_width_setosa, max_sepal_width_setosa, test.loc[i][j])
      fuzzy_rank_versicolor = getscore(min_sepal_width_versicolor, max_sepal_width_versicolor, test.loc[i][j])
      fuzzy_rank_virginica = getscore(min_sepal_width_virginica, max_sepal_width_virginica, test.loc[i][j])
    if ( j==2):
      fuzzy_rank_setosa = getscore(min_petal_length_setosa, max_petal_length_setosa, test.loc[i][j])
      fuzzy_rank_versicolor = getscore(min_petal_length_versicolor, max_petal_length_versicolor, test.loc[i][j])
      fuzzy_rank_virginica = getscore(min_petal_length_virginica, max_petal_length_virginica, test.loc[i][j])
    if (j==3):
      fuzzy_rank_setosa = getscore(min_petal_width_setosa, max_petal_width_setosa, test.loc[i][j])
      fuzzy_rank_versicolor = getscore(min_petal_width_versicolor, max_petal_width_versicolor, test.loc[i][j])
      fuzzy_rank_virginica = getscore(min_petal_width_virginica, max_petal_width_virginica, test.loc[i][j])
    print([fuzzy_rank_setosa,fuzzy_rank_versicolor,fuzzy_rank_virginica], end=" || ")
    total_fuzzy_sum = fuzzy_rank_setosa + fuzzy_rank_versicolor + fuzzy_rank_virginica
    if total_fuzzy_sum == 0.0:
      total_fuzzy_sum = 0.1
    confidence_factor_setosa = fuzzy_rank_setosa/total_fuzzy_sum
    confidence_factor_versicolor = fuzzy_rank_versicolor/total_fuzzy_sum
    confidence_factor_virginica =  fuzzy_rank_virginica/total_fuzzy_sum
    print([confidence_factor_setosa,confidence_factor_versicolor,confidence_factor_virginica])
    arr.append([confidence_factor_setosa,confidence_factor_versicolor,confidence_factor_virginica])
  print('')
  confidence_factor_array.append(arr)
```

#Generation of Modifed Fuzzy Rank Matrix (Fuzzy rank for a class for a particular feature)

```python
import math
modified_fuzzy_rank_array = []

for dataitem in confidence_factor_array:
  f_arr=[]
  for feature in dataitem:
    arr=[]
    for confidence_factor in feature:
      modified_rank = 1-math.exp(((confidence_factor-1.0)*(confidence_factor-1.0))/(-2.0))
      arr.append(modified_rank)
    f_arr.append(arr)
  modified_fuzzy_rank_array.append(f_arr)
modified_fuzzy_rank_array
```

```python
modified_fuzzy_rank_data = modified_fuzzy_rank_array[0]
modified_fuzzy_rank_data
```

#Implementation of the prediction part of CFGaussFRLF
##Calculation for the First Data Point

```python
RSc = {}
CFSc = {}
FSc = {}

for i in range(0, len(modified_fuzzy_rank_data)):
  featured_based = modified_fuzzy_rank_data[i]
  indices = sorted(range(len(featured_based)), key=lambda x: featured_based[x])[0:2]
  for j in range(0, len(featured_based)):
    if j in indices:
      try:
        RSc[j] += featured_based[j]
      except:
        RSc[j] = featured_based[j]
      try:
        CFSc[j] += (1.0/4)*confidence_factor_array[0][i][j]
      except:
        CFSc[j] = (1.0/4)*confidence_factor_array[0][i][j]
    else:
      try:
        RSc[j] += 1.0
      except:
        RSc[j] = 1.0
      try:
        CFSc[j] += (1.0/4)*1.0
      except:
        CFSc[j] = (1.0/4)*1.0
  print(indices)
CFSc_keys = [*CFSc]

for key in CFSc_keys:
  CFSc[key] = (1 - CFSc[key])
print(RSc)
print(CFSc)
valid_keys = list(set([*RSc]) & set([*CFSc]))
for key in valid_keys:
  FSc[key]= RSc[key]* CFSc[key]
print(FSc)

predicted_class = min(FSc, key=FSc.get)
predicted_class
```
